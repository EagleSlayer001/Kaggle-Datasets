```python
import requests
from bs4 import BeautifulSoup
```

```python
response= requests.get('https://www.google.com')
```

```python
content=response.text
```

```python
print(content)
```

```python
Soup=BeautifulSoup(content,'html.parser')
for i in Soup.find_all('a'):
  print(i.get('href'))
```
